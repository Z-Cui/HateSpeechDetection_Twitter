```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

names=['0', '1']

fig = plt.figure(figsize = (15,5)) # width x height
ax1 = fig.add_subplot(1, 4, 1) # row, column, position
ax2 = fig.add_subplot(1, 4, 2)
ax3 = fig.add_subplot(1, 4, 3)
ax4 = fig.add_subplot(1, 4, 4)
ax1.set_title("2-layer LSTM - Unbalanced")
ax2.set_title("BERT - Unbalanced")
ax3.set_title("BERT - Balanced")
ax4.set_title("BERT - Balanced & Augmented")

matrix_proportions = np.array([0.71, 0.29,0.29, 0.71]) # lstm
matrix_proportions = matrix_proportions.reshape(2,2)
confusion_df = pd.DataFrame(matrix_proportions, index=names,columns=names)
sns.heatmap(confusion_df,ax=ax1,annot=True,annot_kws={"size": 20},cmap='gist_gray_r',cbar=False, square=True,fmt='.2f')

matrix_proportions = np.array([0.99, 0.01,0.27, 0.73]) # unbalanced
matrix_proportions = matrix_proportions.reshape(2,2)
confusion_df = pd.DataFrame(matrix_proportions, index=names,columns=names)
sns.heatmap(confusion_df,ax=ax2,annot=True,annot_kws={"size": 20},cmap='gist_gray_r',cbar=False, square=True,fmt='.2f')

matrix_proportions = np.array([0.92, 0.08,0.09, 0.91]) # balanced
matrix_proportions = matrix_proportions.reshape(2,2)
confusion_df = pd.DataFrame(matrix_proportions, index=names,columns=names)
sns.heatmap(confusion_df,ax=ax3,annot=True,annot_kws={"size": 20},cmap='gist_gray_r',cbar=False, square=True,fmt='.2f')

matrix_proportions = np.array([0.94, 0.06,0.05, 0.95]) # balanced & augmented
matrix_proportions = matrix_proportions.reshape(2,2)
confusion_df = pd.DataFrame(matrix_proportions, index=names,columns=names)
sns.heatmap(confusion_df,ax=ax4,annot=True,annot_kws={"size": 20},cmap='gist_gray_r',cbar=False, square=True,fmt='.2f')

fig.show
```
